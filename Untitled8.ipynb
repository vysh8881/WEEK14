# data preprocessing

```python
#checking null values
X.isna().sum()
```

```python
#checking outliers
X.plot(kind='box', subplots=True, layout=(4,4), figsize=(15,15))
plt.title("Outlier Visualization ")
plt.show()
```

```python
#handling outliers
def remove_outlier(col):
    sorted(col)
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range
```

```python
for column in X.columns:
    lr,ur=remove_outlier(X[column])
    X[column]=np.where(X[column]>ur,ur,X[column])
    X[column]=np.where(X[column]<lr,lr,X[column])
```

```python
X.plot(kind='box', subplots=True, layout=(4,4), figsize=(15,15))
plt.title(" After removing Outlier  ")
plt.show()
```

```python
#standardScaler
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df=scaler.fit_transform(X)
df=pd.DataFrame(df,columns=X.columns)

df.describe()
```

# Kmeans clustering

```python
#finding the optimum no:of clusters in kmeans
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans= KMeans(n_clusters=i, init= 'k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    print(kmeans.inertia_)
plt.plot(range(1,11),wcss,'-o')
plt.title("The elbow method")
plt.xlabel("Number of clusters")
plt.ylabel("wcss")
plt.show()
```

```python
# create the kmeans model with the optimum number of clusters 3
kmeans=KMeans(n_clusters=3,init='k-means++',random_state=42)
y_kmeans=kmeans.fit_predict(X)
type(X)
y_kmeans
```

```python
X['cluster'] = y_kmeans
```

```python
l=['child_mort', 'exports', 'health', 'imports', 'income',
       'inflation', 'life_expec', 'total_fer', 'gdpp']
plt.figure(figsize=(20,20))
for i in range(len(l)):
    plt.subplot(3,3,i+1)
    sns.boxplot(X.cluster,X[l[i]])
    plt.title(l[i])
```

```python
cluster_values=X.groupby('cluster').mean()
cluster_values['Freq']=X['cluster'].value_counts().sort_index()
cluster_values
```
